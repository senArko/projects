```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
sns.set_style("darkgrid")
matplotlib.rcParams["figure.figsize"] = (10,5)
%matplotlib inline
```

```python
data = pd.read_csv("Stores.csv")
```

```python
data
```

```python
# Atores with area greater than the average area

m = round(data["Store_Area"].mean(),0)
```

```python
a = data[data["Store_Area"] > m]
```

```python
a
```

```python
# Average sales of the stores whose area is more than the average area.

a["Store_Sales"].mean()
```

```python
# Stores whos area is less than the average area

less = data[data["Store_Area"] < m]
```

```python
less
```

```python
less["Store_Sales"].mean()
```

After comparing the average sales of the stores whose area is more and less than the average area we can conclude that the store area can effect the overall sales of the stores.

```python
# Items available in the stores whose area is more than the average

a["Items_Available"].mean()
```

```python
# Items available in the stores where the area is less than the average

less["Items_Available"].mean()
```

The largeer the area of the store the more item can be kept in the store.

```python
df = pd.read_csv("Sample - Superstore.csv", encoding = "latin-1")
```

```python
df
```

```python
# Checking for null values

df.isnull().sum()
```

```python
# Checking the shape of the data

df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing the data type

df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Date"] = df["Order Date"].apply(lambda x: x.day)
```

```python
df.head()
```

```python
# Date wise total orders

df = df.sort_values(by = "Date")
```

```python
df.groupby("Date")["Order ID"].count().plot()
```

```python
# Count of shipping methods

ax = sns.countplot(x = "Ship Mode", data = df, palette = "viridis")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Count of order segment

ax = sns.countplot(x = "Segment", data = df, palette = "Set2")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# Region-wise total sales

ax = round(df.groupby("Region")["Sales"].sum(),2).plot(kind = "bar")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Region wise total profit

profit = df.groupby("Region")["Profit"].sum()
```

```python
ax = sns.barplot(x = profit.index, y = profit.values)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 cities with the highest sales

city = df.groupby("City")["Sales"].sum().sort_values(ascending = False).head(10)
```

```python
ax = sns.barplot(x = city.index, y = city.values)

plt.xticks(rotation = 90)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df["Order Date"].dt.year
```

```python
# Creating a year column

df["Year"] = df["Order Date"].dt.year
```

```python
df.head()
```

```python
# Yearly profit trends

sns.lineplot(x = df["Year"], y = df["Profit"])
```

```python
df.columns
```

```python
# Category and sub category wise quantity ordered

df.groupby(["Category","Sub-Category"])["Quantity"].sum().unstack().plot(kind = "bar", stacked = False)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Yearly sales trends

df.groupby("Year")["Sales"].sum().plot()
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df["Order_to_Ship"] = df["Ship Date"] - df["Order Date"]
```

```python
df.head()
```

```python
# Shipping time

ax = sns.countplot(x = "Order_to_Ship", data = df, palette = "viridis")

plt.xticks(rotation = 90)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# Average discount per product category

df.groupby("Category")["Discount"].mean().plot(kind = "bar")
```

```python
# Correlation between sales and profit

sns.regplot(x = df["Sales"], y = df["Profit"])
```
