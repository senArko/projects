ANALYSIS OF FOOTBALL ATTENDENCE OF AUDIENCE

```python
# Dropping extra columns

df1 = df1.drop("Unnamed: 0",axis = 1)
```

```python
df1.head()
```

```python
df1.drop([22],axis = 0,inplace = True)
```

```python
df1
```

```python
# Checking for null values

df1.isnull().sum()
```

```python
# Count of hosts

host = df1["Hosts"].value_counts()

host
```

```python
host.plot(kind = "bar")
plt.title("Count of host")
plt.xlabel("Host")
plt.ylabel("Count")
```

```python
# Year -wise total attendence

sns.lineplot(df1["Year"],df1["Total_Attendance"])

plt.title("Yesr-wise total attendence")
```

```python
# Year-wise average attendence


sns.lineplot(df1["Year"], df1["Average_Attendance"])

plt.title("Year-wise average attendence")
```

```python
df1.head()
```

```python
# Total matches help per venue

venue = df1["Venue"].value_counts()
```

```python
venue
```

```python
venue.plot(kind = "barh")
```

```python
# Matches per host

sns.barplot(df1["Hosts"],df1["Matches"],ci = False)

plt.xticks(rotation = 90)

plt.title("Matches played in each host country")
```
