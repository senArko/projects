```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
sns.set_style("darkgrid")

from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (10,5)
```

```python
df = pd.read_csv("Fixture T20 World Cup 2024.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Checking for null values

df.isnull().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Day of week wise total matches

sns.countplot("Day", data = df, palette = "viridis")
```

```python
# Group-wise total matches

group = df["Group"].value_counts()
```

```python
sns.barplot(x = group.index,y = group.values)
```

```python
# Venue-wise total matches

sns.countplot("Venue", data = df, palette = "Set2")

plt.xticks(rotation = 90)

plt.title("Venue-wise total matches")
```

```python
# Country-wise total matches

sns.countplot("Country", data = df, palette = "muted")

plt.title("Country-wise total matches")
```
