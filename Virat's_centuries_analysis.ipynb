```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
sns.set_style("darkgrid")

from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (10,5)
```

```python
df = pd.read_csv("71 Centuries of Virat Kohli.csv")
```

```python
df.head()
```

```python
# Missing value treatment


df["Strike Rate"].unique()
```

```python
sns.heatmap(df.isnull(),cmap ="viridis", cbar = False)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["Strike Rate"] = df["Strike Rate"].fillna(round(df["Strike Rate"].mean(),2))
```

```python
df.drop("Unnamed: 14",axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing the Date column Date Time datatype

df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Comparing Out and Not out centuries

out_not_out = df["Out/Not Out"].value_counts()
```

```python
plt.pie(out_not_out,labels = out_not_out.index,autopct="%.2f%%",explode = [0.1,0.1])
```

```python
# Renaming Column

df.rename({"Column1":"City"},axis = 1,inplace = True)
```

```python
df.head()
```

```python
# Against which team Virat has scored the maximum centuries

ax=sns.countplot("Against",data = df, palette = "viridis")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
         fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
df["Batting Order"].unique()
```

```python
# Which batting order did Virat bat the most

ax=sns.countplot("Batting Order", data = df, palette = "Set1")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
         fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
# Top 5 venues Virat has scored the most centuries on

venue = df["Venue"].value_counts().sort_values(ascending = False).head(5)

venue
```

```python
ax=sns.barplot(x = venue.index, y = venue.values, color = "blue")
plt.xticks(rotation = 90)
plt.xlabel("Venue")
plt.ylabel("Total Centuries")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
         fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
# Funtion that will return total wins against different teams

def centuries(name):
    total_centuries = df[df.Against == name]["Against"].count()
    return total_centuries

name = input("Enter team name: ")
centuries = centuries(name)
print(centuries)
```
