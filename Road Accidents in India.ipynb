```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
sns.set_style("darkgrid")
matplotlib.rcParams["figure.figsize"] = (10,5)
%matplotlib inline
```

```python
data = pd.read_csv("RA2019_A2.csv")
```

```python
data
```

```python
# Checking for null values:

data.isnull().sum()
```

```python
# Dropping the null values

data.dropna(inplace = True)
```

```python
data.set_index("States/UTs", inplace = True)
```

```python
data.head()
```

```python
data.columns
```

```python
# Renaming lenghty column names:

data = data.rename(columns={
    "State/UT-Wise Total Number of Road Accidents during 2016": "Total_Acidents(2016)",
    "State/UT-Wise Total Number of Road Accidents during 2017": "Total_Accidents(2017)",
    "State/UT-Wise Total Number of Road Accidents during 2018": "Total_Accidents(2018)",
    "State/UT-Wise Total Number of Road Accidents during 2019 - Numbers": "Total_Accidents(2019)",
    "State/UT-Wise Total Number of Road Accidents during 2019 - Rank": "Total_Accidents_2019(Rank)",
    "Share of States/UTs in Total Number of Road Accidents - 2016": "Share_in_total(2016)",
    "Share of States/UTs in Total Number of Road Accidents - 2017": "Share_in_total(2017)",
    "Share of States/UTs in Total Number of Road Accidents - 2018": "Share_in_total(2018)",
    "Share of States/UTs in Total Number of Road Accidents - 2019": "Share_in_total(2019)",
    "Total Number of Accidents Per Lakh Population - 2016": "Accidents/lakh_population(2016)",
    "Total Number of Accidents Per Lakh Population - 2017": "Accidents/lakh_population(2017)",
    "Total Number of Accidents Per Lakh Population - 2018": "Accidents/lakh_population(2018)",
    "Total Number of Accidents Per Lakh Population - 2019": "Accidents/lakh_population(2019)",
    "Total Number of  Road Accidents per 10,000 Vehicles - 2016": "Accidents/10000_vehicles(2016)",
    "Total Number of  Road Accidents per 10,000 Vehicles - 2017": "Accidents/10000_vehicles(2017)",
    "Total Number of  Road Accidents per 10,000 Vehicles - 2018": "Accidents/10000_vehicles(2018)",
    "Total Number of  Road Accidents per 10,000 Vehicles - 2019": "Accidents/10000_vehicles(2019)",
    "Total Number of  Road Accidents per 10,000 Km of Roads - 2016": "Accidents/10000km_road(2016)",
    "Total Number of  Road Accidents per 10,000 Km of Roads - 2017": "Accidents/10000km_road(2017)",
    "Total Number of  Road Accidents per 10,000 Km of Roads - 2018(P)": "Accidents/10000km_road(2018)"

})
```

```python
data.columns
```

```python
data.head()
```

```python
# Year-wise total accidents in different state

sns.lineplot(x = data.index, y = data["Total_Acidents(2016)"], label = 2016)
sns.lineplot(x = data.index, y = data["Total_Accidents(2017)"], label = 2017)
sns.lineplot(x = data.index, y = data["Total_Accidents(2018)"], label = 2018)
sns.lineplot(x = data.index, y = data["Total_Accidents(2019)"], label = 2019)

plt.xticks(rotation = 90)

plt.title("Year-wise total accidents")
```

```python
# Year-wise share in total accidents 

sns.lineplot(x = data.index, y = data["Share_in_total(2016)"], label = 2016)
sns.lineplot(x = data.index, y = data["Share_in_total(2017)"], label = 2017)
sns.lineplot(x = data.index, y = data["Share_in_total(2018)"], label = 2018)
sns.lineplot(x = data.index, y = data["Share_in_total(2019)"], label = 2019)



plt.xticks(rotation =90)

plt.title("Share in total")
```

```python
data.columns
```

```python
# Accidents per 100000 of population

round(data.groupby(level = 0)[["Accidents/lakh_population(2016)","Accidents/lakh_population(2017)","Accidents/lakh_population(2018)","Accidents/lakh_population(2019)"]].sum(),0).plot(kind = "bar", stacked = False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title("Accidents per 100000 of population")
```

```python
# Accidents per 10000 vehicles


round(data.groupby(level = 0)[["Accidents/10000_vehicles(2016)","Accidents/10000_vehicles(2017)","Accidents/10000_vehicles(2018)","Accidents/10000_vehicles(2019)"]].sum(),0).plot(kind = "bar",stacked = False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title("Accidents per 10000 vehicles")
```

```python
# Accidents per 10000 km of raod

round(data.groupby(level = 0)[["Accidents/10000km_road(2016)","Accidents/10000km_road(2017)","Accidents/10000km_road(2018)"]].sum(),0).plot(kind = "bar", stacked = False)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title("Accidents per 10000 km of road")
```
