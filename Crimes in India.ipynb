```python
data["Total Registered Cases"] = data["2014 - Cases registered"] + data["2015 - Cases registered"] + data["2016 - Cases registered"]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data.index,data["Total Registered Cases"])
plt.xticks(rotation = 85)
plt.xlabel("State/UT")
plt.title("Total Registered Cases 2014-2016")

```

```python
data.head()
```

```python
data["Total Rapes"] = data["Rape - 2015"] + data["Rape - 2016"] + data["2014 - Total rape Cases"]
```

```python
def count(i):
    if i <= 150:
        return "Low"
    elif i > 150 and i <= 1000:
        return "Moderate"
    else:
        return "High"
```

```python
data["Rape Count"] = data["Total Rapes"].apply(count)
```

```python
ax = sns.countplot(x = "Rape Count", data = data, palette = "Set1")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
        fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
plt.boxplot(data["Total Rapes"], showmeans= True)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x = data.index, y = (data["Assaults (molestation) - 2015"]/data["2015 - Cases registered"])*100,label = 2015)
sns.lineplot(x = data.index, y = (data["Assaults (molestation) - 2016"]/data["2016 - Cases registered"])*100,label = 2016)
plt.xticks(rotation = 90)
plt.xlabel("State/UT")
plt.ylabel("% Molestation")
plt.title("% 2015-2016 Molestation")
```

```python
sns.barplot(x = data.index, y = (data["Murder (women) - 2015"]/data["2015 - Cases registered"])*100,label = 2015)
sns.barplot(x = data.index, y = (data["Murder (women) - 2016"]/data["2016 - Cases registered"])*100,label = 2016)
plt.xticks(rotation = 90)
plt.xlabel("State/UT")
plt.ylabel("% Murder")
plt.title("% Murder 2015-2016")
```

```python
data["Rape %"] = round((data["2014 - Total rape Cases"]+data["Rape - 2015"]+
                  data["Rape - 2016"])/(data["2014 - Cases registered"]+data["2015 - Cases registered"]+data["2016 - Cases registered"])*100,2)
```

```python
data.head()
```

```python
data.drop(columns=["Rape %"],inplace = True)
```

```python
def safe(i):
    if i <5:
        return "Safe"
    elif i>=5 and i<=10:
        return "Moderately Safe"
    else:
        return "Unsafe"
```

```python
data["How Safe"] = data["Rape %"].apply(safe)
```

```python
ax = sns.countplot(x = "How Safe", data = data)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
        fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')
    
plt.title("Safe or Unsafe")
```
